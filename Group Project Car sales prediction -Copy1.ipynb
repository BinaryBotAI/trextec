## Data Cleaning

```python
# Fixing column names
data.columns = ['Make', 'Model', 'Year', 'Fuel_Type', 'HP',
'Transmission', 'Drive_Wheels', 'Vehicle_Size', 'Vehicle_Style', 'Popularity',
'MSRP', 'Price(Ksh)', 'Engine_CC']
```

```python
# Drop duplicates
data.drop_duplicates(inplace=True)
```

```python
# Fixing the datatype for Year feature
data['Year'] = pd.to_datetime(data['Year'], errors='coerce')
data['Year'] = data['Year'].dt.year
```

```python
# Filling missing values with the median for numerical features
num_cols = ['HP', 'Popularity', 'MSRP', 'Price(Ksh)', 'Engine_CC']
for col in num_cols:
    data[col].fillna(data[col].median(), inplace=True)
```

```python
# Dropping missing values for categorical features
cat_cols = ['Make', 'Model','MSRP', 'Fuel_Type', 'Transmission', 'Drive_Wheels', 'Vehicle_Size', 'Vehicle_Style']
data.dropna(subset=cat_cols, inplace=True)
```

This code drops the columns that are not needed for predicting the car price. These columns include 'Engine Fuel Type', 'Drive_Wheels', 'Vehicle Size', 'Vehicle Style', and 'Popularity'.

```python
# Drop unnecessary column(s) 
data = data.drop('MSRP', axis=1)
```

```python
# Checking for null values again
data.isnull().sum()
```

```python
# Converting categorical features to numerical using LabelEncoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cat_cols = ['Make', 'Model', 'Fuel_Type', 'Transmission', 'Drive_Wheels', 'Vehicle_Size', 'Vehicle_Style']

for col in cat_cols:
    data[col] = le.fit_transform(data[col])
```

# Machine Learning

## Data Preprocessing

```python
# Splitting the dataset into features and labels

X = data[['Make', 'Model', 'Fuel_Type', 'Transmission', 'Drive_Wheels', 'Vehicle_Size', 'Vehicle_Style']]

# Selecting label column
y = data['Price(Ksh)']
```

```python
# Scaling numerical features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Splitting data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

## Model Training


```python
# Training a linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Standardize the training data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Replace missing values with the median of each column
imputer = SimpleImputer(strategy='median')
X_train_scaled = imputer.fit_transform(X_train_scaled)

# Train the linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled, y_train)
```

```python
# final dataset used in training 
data.head ()
```

## Model Evaluation
In this section, we evaluated our model on the testing set to assess its performance. We calculated the mean absolute error, root mean squared error, and R-squared score of the predictions made by the linear regression model.

```python
# Test the model on the testing set
y_pred = lin_reg.predict(X_test)
```

```python
# Evaluating the model using mean absolute error
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error: {mae:.2f}')
```

```python
# Evaluating the model using root mean squared error
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'Root Mean Squared Error: {rmse:.2f}')
```

```python
# Evaluating the model using R-squared score
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2:.2f}')
```

The mean absolute error was found to be 493,676.89, which means that on average, the model's predictions deviated from the actual values by $493,676.89. The root mean squared error was found to be 729,654.75, which represents the square root of the average squared difference between the predicted and actual values. Finally, the R-squared score was 0.99, indicating that 99% of the variance in the dependent variable was explained by the independent variable.
